```python
# https://www.researchgate.net/publication/313795371_Optimization_of_P_Median_Problem_in_Python_Using_PuLP_Package

from pulp import *
import pandas as pd
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
from kneed import KneeLocator
import folium
from folium import plugins
from matplotlib import rc
```

```python
rc('font', family='AppleGothic')

plt.rcParams['axes.unicode_minus'] = False
```

```python
dongs = ['', '1', '', '']

elbow_points = []

for dong in dongs:
    file_name_d = './ /' + dong + '__.xlsx'
    file_name_w = './ /_' + dong + '__.xlsx'
    
    d = pd.read_excel(file_name_d).iloc[:, 1:] # Distance matrix (Index Column )
    w = pd.read_excel(file_name_w).iloc[:, 0] #   (  )
    
    for j in range(d.shape[1]):
        for i in range(d.shape[0]):
            d.iloc[i, j] = d.iloc[i, j].replace(' km', '')
        d.iloc[:, j] = d.iloc[:, j].astype(float)

    d = d.values
    w = w.values

    I = [i for i in range(d.shape[0])] #  index array
    J = [j for j in range(d.shape[1])] #  index array
    #    

    answers = []
    zs = []

    for p in range(1, len(J) + 1):
        prob = LpProblem('P_Median', LpMinimize)
        a = LpVariable.dicts('a_blocks_%s_publics_%s', (I, J), cat = 'Binary')
        x = LpVariable.dict('X_%s', J, cat = 'Binary')
        prob += sum(sum(w[i] * d[i][j] * a[i][j] for j in J) for i in I)
        prob += lpSum([x[j] for j in J]) == p
        for i in I:
            prob += lpSum([a[i][j] for j in J]) == 1
        for j in J:
            for i in I:
                prob += a[i][j] <= x[j]
                
        prob.solve()
        
        a_soln = np.array([[a[i][j].varValue for j in J] for i in I])
        x_soln = np.array([x[j].varValue for j in J])
        
        answers.append(x_soln)
        zs.append(value(prob.objective))
        
    plt.figure(figsize = (7, 4))
    plt.plot(range(1, len(J) + 1), zs)
    plt.xlabel('  ()')
    plt.ylabel('   ( * km)')
    plt.title(dong + ' P-Median  ')
    plt.savefig(dong + '_P-Median')
    plt.show()
    
    kn = KneeLocator(range(1, len(J) + 1), zs, curve = 'convex', direction='decreasing')
    elbow_points.append(answers[kn.knee - 1])
```

```python
CENTER = [37.5648406, 126.9773030] #  
m = folium.Map(location=CENTER, zoom_start=11)

for i, dong in enumerate(dongs):
    print(dong)
    print('\n')
    
    public = pd.read_excel('.//' + dong + '.xlsx')
    block = pd.read_excel('./ /_' + dong + '__.xlsx')
    
    result = public.loc[(np.where(elbow_points[i] == 1)[0]), :]
    
    for row_index, row in enumerate(public.itertuples()):
        if row_index in (np.where(elbow_points[i] == 1)[0]):
            print(public.at[row_index, ''])
            location = [float(public.at[row_index, '']), float(public.at[row_index, ''])]
            folium.Marker(location, tooltip='<b>{}</b>'.format(public.at[row_index, ''])).add_to(m)

    location_data = block[['', '', ' ']].values[:len(block)].tolist()
    plugins.HeatMap(location_data).add_to(m)
m.save('__P-Median.html')
```
