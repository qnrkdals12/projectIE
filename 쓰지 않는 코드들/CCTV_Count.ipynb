```python
import pandas as pd
from tqdm import tqdm
```

```python
import os

files = []

for i in os.listdir():
    if i.endswith('.csv'):
        files.append(i)
```

```python
files
```

```python
data = pd.read_csv(files[0], encoding = 'cp949')
```

```python
data
```

```python
names = data['행정동'].unique()
print(len(names))
```

```python
count = pd.DataFrame()
count['행정동'] = names
count['CCTV 수'] = ''
```

```python
for i in count.itertuples():
    number = 0
    for j in data.itertuples():
        if i[1] == j[-1]:
            number += int(j[6])
    count.at[i.Index, 'CCTV 수'] = number
```

```python
count
```

```python
count.to_csv('The number of CCTV.csv', encoding = 'cp949')
```
