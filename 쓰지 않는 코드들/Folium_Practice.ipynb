```python
# import station

import folium
import pandas as pd
import os
from folium import plugins
import json
import numpy as np
```

```python
my_map = folium.Map(location=[37.5756713, 127.0329137], zoom_start=14) # 지도 생성 변수
```

```python
my_map2 = folium.Map(location=[37.5756713, 127.0329137], zoom_start=14, tiles='Stamen Terrain') # 지도 이미지 변경 (tiles=)
```

```python
for i in os.listdir():
    if i.endswith('.xlsx'):
        publics = pd.read_excel(i).iloc[:,1:3] # Column 1 (String), Column 2 (List)
publics
```

```python
for row_index, row in enumerate(publics.itertuples()):
    publics.iloc[row_index, -1] = row[-1].replace('(', '').replace(')', '').split(', ')
publics
```

```python
folium.Marker(publics.iloc[0,1]).add_to(my_map) # 마커 표시
```

```python
# 클릭했을 때 정보 표시 (popup=)
folium.Marker(publics.iloc[1,1], popup='<b>publics</b>').add_to(my_map)
```

```python
# 마우스를 올렸을 때 정보 표시 (tooltip=)
folium.Marker(publics.iloc[2,1], popup='<b>publics2</b>', tooltip='<i>Where?</i>').add_to(my_map)
```

```python
folium.Circle(location=publics.iloc[3,1],
             radius=500,
             color='#eb9e34',
             fill_color='red',
             popup='Circle popup',
             tooltip='Circle tooltip').add_to(my_map)
```

```python
blocks = pd.read_excel('좌표_최종_용신동.xlsx')
```

```python
blocks.info()
```

```python
blocks['좌표'] = ''

for index, value in enumerate(blocks.itertuples()):
    blocks.at[index, '좌표'] = [float(value[-2]), float(value[-3])]
blocks
```

```python
location_data = np.array([blocks['위도'], blocks['경도'], blocks['택배 착지량']])
```

```python
for index, value in enumerate(blocks.itertuples()):
    max_bc_count = int(value[0])
    folium.Circle(location=value[-1], radius=max_bc_count, color='red', fill=True, fill_opacity=1, popup=max_bc_count).add_to(my_map)
```

```python
plugins.HeatMap(location_data).add_to(my_map)
```
