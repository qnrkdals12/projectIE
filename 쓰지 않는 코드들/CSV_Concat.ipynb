```python
import os

files = []

for i in os.listdir():
    if i.endswith('.csv'):
        files.append(i)
print(files)
```

```python
import pandas as pd

first = pd.read_csv(files[2])
first.drop('Unnamed: 0', axis = 1, inplace = True)
second = pd.read_csv(files[3], encoding = 'cp949')
second.drop('Unnamed: 0', axis = 1, inplace = True)
```

```python
result = pd.merge(first, second, how = 'outer', on = '동' )
```

```python
result.to_excel('result.xlsx', encoding = 'cp949')
```
